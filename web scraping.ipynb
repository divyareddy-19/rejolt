```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.flipkart.com/computers/laptops/~dell-gaming-laptops21/pr?sid=6bo%2Cb5g&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIkRlbGwgR2FtaW5nIExhcHRvcHMiXSwidmFsdWVUeXBlIjoiTVVMVElfVkFMVUVEIn19fX19&wid=9.productCard.PMU_V2_5'
```

```python
req=requests.get(url)
```

```python
content=BeautifulSoup(req.content,'html.parser')
print(content)
```

```python
print(content)
```

```python
data=content.find_all('div',{'class':'_2kHMtA'})
```

```python
print(data)
```

```python
links=[]
lap_name=[]
start_link="https://www.flipkart.com/"
```

```python
for items in data:
    rest_link=items.find('a')['href']
    name=items.find('div',attrs={'class':'_4rR01T'})
    lap_name.append(name.text)
    links.append(start_link+rest_link)
```

```python
print(lap_name[0])
print(links[0])
```

```python
dataframe={'laptop_names':lap_name,'links':links}
Final_dataframe=pd.DataFrame(dataframe)
print(Final_dataframe)
```

```python
Final_dataframe.to_csv('fk_data_url.csv')
```
