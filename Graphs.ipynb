```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cosmetics=['LAKME','SUGAR','COLOBAR','MAYBELINE','REVLON']
data=[45,56,33,30,21]
fig=plt.figure(figsize=(11,6))
plt.pie(data,labels=cosmetics)
plt.title("pie chart")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data={'Fogg':40,'Park avenue':60,'Ossum':20,'Blue label':50}
Perfumes=list(data.keys())
values=list(data.values())
fig=plt.figure(figsize=(9,7))
plt.bar(Perfumes,values,color='blue',width=0.4)
plt.xlabel('Perfumes')
plt.ylabel('No of customers purchased')
plt.title('Bar Chart')
plt.show()
```
